```python
class NegativeException(Exception):
    "negatieve number is not valid"
    def __init__(self,arg):
        self.arg = arg
try:
    a = int(input("Enter Number :"))
    if a > 0:
        print(a)
    else:
        raise NegativeException("Negative number")
except NegativeException as e:
    print(e.arg)
```

```python
class Area:
    def __init__(self,r):
        self.r = r

    def Area(self):
        return (2*3.14*r*r)
r = 10
a = Area(r)
area = a.Area()
print(area)
```
