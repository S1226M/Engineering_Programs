### ✅ Task 13 — Boolean Masking
Given:

1. Create a boolean mask for all scores **greater than 60**.
2. Use the mask to extract only those scores.
3. Print the mask and the filtered scores.

````python
# ✏️ Write your answer here
scores = np.array([45, 78, 90, 32, 55, 88, 67, 23, 95, 40])
arr = scores > 60
print(arr)
print(scores[arr])
````

### ✅ Task 15 — Stacking Arrays
Given:
```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```
1. Stack them **vertically** using `np.vstack`. Print the result.
2. Stack them **horizontally** using `np.hstack`. Print the result.
3. Stack them using `np.column_stack`. Print the result.

````python
# ✏️ Write your answer here
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
s = np.vstack(a)
m = np.vstack(b)
print(s,m)

i = np.hstack(a)
t = np.hstack(b)
print(i,t)

m = np.column_stack(a)
a = np.column_stack(b)
print(m,a)
````
