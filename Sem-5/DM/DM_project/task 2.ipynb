# Step 1: Load Data

```python
import pandas as pd
```

```python
data = pd.read_csv("OnlineRetail.csv", encoding='latin1')
```

```python
data
```

# Step 2: Drop Missing Invoices or Descriptions

```python
data = data.dropna(subset=['InvoiceNo', 'Description'])
```

```python
data
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
data = data[~data['InvoiceNo'].astype(str).str.lower().str.startswith('c')]
```

```python
data
```

# Step 4: Keep Only Positive Quantities

```python
data = data[data['Quantity'] > 0]
```

```python
data
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
data['Description'].str.strip()
```

```python
data
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
data = data[data['Country'] == 'United Kingdom']
```

```python
data
```

# Step 7: Remove Duplicates

```python
data = data.drop_duplicates()
data
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']
data
```

# Step 9 : Check for Top 10 Selling Products

```python
top = data.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
top
```

# Step 10 : Identify Most Active Customers

```python
mostactive= data.groupby('CustomerID')['InvoiceNo'].count().sort_values(ascending=False).head(10)
mostactive
```
