```python
import numpy as np
```

```python
# Output 1 if positive, 0 if negative

def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0
print("Setup completed..")
```

## Part 1. The OR Gate

```python
#inputs
X_or = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

#targets
y_or = np.array([0, 1, 1, 1])
```

## Step 2. Initialize weights

```python
weights = np.array([0, 0])
bias = 0
lr = 0.5  #Learning rate

print(f"Initial weights: {weights}")
print(f"Initial Bias: {bias}")
```

## Step 3. Manual Calculation

```python
# Example Input: (0, 1)
input_sample = X_or[1]
target_value = y_or[1]

# 1. Weights Sum (Dot product)
z = np.dot(input_sample, weights) + bias
print(f"Math: ({input_sample[0]} * {weights[0]}) + ({input_sample[1]} * {weights[1]}) + {bias}")
print(f"Sum (z): {z:.2f}")

# 2. Activation
prediction = step_function(z)
print(f"Prediction: {prediction}")

# 3. Error
error = target_value - prediction
print(f"Error: {error}")
```

```python
epochs = 5

print ("---Training---")

for epoch in range(epochs):
    print(f"\nEPOCH--> {epoch + 1}")

    for i in range(4):
        x = X_or[i]
        y = y_or[i]

        # Predict 
        z = np.dot(x, weights) + bias
        pred = step_function(z)

        #Error
        error = y - pred


        # Learn (Update Weights)
        if error != 0:
            weights = weights + (lr * error * x)
            bias = bias + (lr * error * 1)  # always ONE
            print(f"--> Mistake on {x}. Expected {y}. got {pred}. ")
            print(f"--> Updated Weights: {weights} Bias: {bias:.2f} ")

        else:
            print(f" ✅ Correct on {x}")

print("\nFinal Weights: ",weights)
print("Final Bias: ",bias)
```
