```python
data = data.drop_duplicates()
data
```

```python
#normal range of height is = 150 – 185 cm (Adults 18 – 60 years)
#normal range of weight is = 50 – 85 kg (Adults 18 – 60 years)
#basic range of ap_high is = 110 – 130
#basic range of ap_low is = Normal: 70 – 85
```

```python
data['age_years'] = (data['age'] / 365.25).astype(int)
```

```python
data
```

```python
data['Pulse_Pressure'] = (data['ap_hi'] - data['ap_lo'])
data
```

```python
data = data[
    (data['ap_hi'] >= 90) & (data['ap_hi'] <= 250) &
    (data['ap_lo'] >= 60) & (data['ap_lo'] <= 200)
]
```

```python
data = data.drop(columns=['age'])
```

```python
data
```

```python
swap_mask = data['ap_lo'] > data['ap_hi']
data.loc[swap_mask, ['ap_hi', 'ap_lo']] = data.loc[swap_mask, ['ap_lo', 'ap_hi']].values
```

```python
data
```

```python
data.drop(columns=['id'], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Columns to check
cols_to_check = ['height', 'weight', 'ap_hi', 'ap_lo', 'age_years', 'Pulse_Pressure']

data_before = data.copy()

# IQR Outlier Removal
for col in cols_to_check:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]

# After-cleaning copy
data_after = data.copy()

# PLOTTING FUNCTION
def plot_boxplots(df, title):
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    fig.suptitle(title, fontsize=18)

    for ax, col in zip(axes.flatten(), cols_to_check):
        sns.boxplot(y=df[col], ax=ax)
        ax.set_title(col)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# SHOW PLOTS
plot_boxplots(data_before, "Before Outlier Removal (IQR)")
plot_boxplots(data_after, "After Outlier Removal (IQR)")
```
