# Decision Tree Classifier

```python
from sklearn.datasets import load_iris 
```

# Import iris data ser using sklearn

```python
df = load_iris()
```

```python
df
```

# Importing the libraries

```python
import pandas as pd
import numpy as np
```

# Divide the data into input and output

```python
x = df['data']
```

```python
y = df['target']
```

```python
y
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Fitting DecisionTreeClassifier   on dataset

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(random_state=42, criterion='entropy', max_depth=100)
```

```python
model.fit(x_train,y_train)
```

# Display Decision Tree

```python
%matplotlib qt
```

```python
tree.plot_tree(model)
```

# Predict the x_test 

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Import diabetes.csv dataset 

```python
data = pd.read_csv('diabetes.csv')
```

```python
data
```

# Check the distribution of the target

```python
data.value_counts('Outcome')
```

# Replace zeros with NaN for specific columns

```python
data = data.replace(0,np.nan)
```

```python
data
```

# Check for missing values

```python
data.isnull().sum()
```

# Fill missing values with median

```python
numeric_cols = data.select_dtypes(include='number').columns

for col in numeric_cols:
    data[col].fillna(data[col].median(), inplace=True)
```

```python
data.isnull().sum()
```

# Visualize Distributions

# Convert Targer data into interger code

```python
data['Outcome'] = data['Outcome'].map({
    'Non Diabetic': 0,
    'Diabetic': 1
})
```

```python
data
```

# Divide the data into input and output

```python
Input = data.drop('Outcome', axis=1)
Output = data['Outcome']
```

```python
Input
```

```python
Output
```

# Splitting the dataset into the Training set and Test set

```python
X_train, X_test, Y_train, Y_test = train_test_split(Input, Output, test_size=0.2, random_state=42)
```

# Create Model

```python
model1 = tree.DecisionTreeClassifier(random_state=42, criterion='entropy', max_depth=100)
```

# Fitting DecisionTreeClassifier   on dataset

```python
model1.fit(X_train,Y_train)
```

# Display Decision Tree

```python
%matplotlib qt
tree.plot_tree(model1)
```

# Predict the x_test 

```python
model1.predict(X_test)
```

# Display Training Accuracy

```python
model1.score(X_train,Y_train)
```

# Display Test Accuracy

```python
model1.score(X_test,Y_test)
```
